```python
def list_str_read(str_list_in):
    list_out = []
    if str_list_in != "[]":
        str_list_in = str_list_in.strip('[]')
        for item in str_list_in.split(','):
            list_out.append(item.strip("' '")) #Note: may not need space here
    return list_out
```
