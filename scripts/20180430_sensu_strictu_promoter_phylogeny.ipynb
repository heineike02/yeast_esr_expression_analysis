```python
#Make a dataframe based on the YGOB genome.tab file
strand_dict = {0: 'Crick', 1: 'Watson'}
spec_fullname_dict = {"Skud": "Skudriavzevii", 
                      "Smik": "Smikatae",
                      "Suva": "Suvarum", 
                      "Kafr": "Kafricana",
                      "Cgla": "Cglabrata"}

spec_lookups = {}
for spec in specs_not_scer:

    genome_lookup_fname = os.path.normpath(data_processing_dir + "promoter_phylogenies/genomes/" + spec_fullname_dict[spec]+"_genome.tab")

    spec_lookup = pd.read_table(genome_lookup_fname,header = None)
    spec_lookup.columns = ['name','strand','start','stop','YGOB on/off','chromosome','short name','coordinates','notes']

    spec_lookup.set_index('name',inplace=True,verify_integrity=True)
    spec_lookups[spec]=spec_lookup

spec_lookups

# These files are ordered lists of the genes in each 
#  of the genomes included in this version of YGOB. 
#  The columns contain, from left to right:

#  - NAME - unambiguous name used to identify the gene
#  - ORIENTATION - 0 = Crick Strand, 1 = Watson Strand
#  - START COORDINATE - 5' start of the co-ordinate range of the feature
#  - STOP COORDINATE - 3' end of the co-ordinate range of the feature
#  - ON/OFF - whether feature is displayed or not in YGOB
#  - CHROMOSOME/CONTIG/SCAFFOLD NUMBER - identifying number of source sequence
#  - SHORT NAME - the shorter name that will appear in the gene box on screen in YGOB
#  - COORDINATES - complete gene co-ordinates with intron/exon annotation and complement tag if appropriate
#  - NOTES - GenBank tags, SGD descriptions and notes for the locus

#  The only exception being the 'Ancestor_genome.tab' file as
#  the ancestral gene order has no underlying sequence and hence
#  no co-ordinates. In this case the columns are:

#  - NAME - unambiguous name used to identify the gene
#  - ORIENTATION - 0 = Crick Strand, 1 = Watson Strand (inferred from homologs)
#  - ON/OFF - whether feature is displayed or not in YGOB
#  - CHROMOSOME NUMBER - identifying number of ancestral chromosome
#  - SHORT NAME - the shorter name that will appear in the gene box on screen in YGOB
```

```python
#For a given S.Cer gene, get the promoter sequences from the sensu strictu species and output into a file. 

#Extract Sensu strictu orthologs from pillars. 

gene_name = 'CYC7'
gene = io_library.SC_orf_lookup_by_name([gene_name])[0]
N_prom = 700

#extract promoter for S.Cer
fname_scer_proms = os.path.normpath(data_processing_dir + 'sc_promoters/promoter_sets/all_sc_promoters.fasta')
scer_proms = SeqIO.index(fname_scer_proms, "fasta")
scer_prom = scer_proms[gene]
prom_records = [scer_prom]

#Lookup orthologs based on the YGOB Pillars dataframe
gene_pos = ''
for pos in ['p1','p2']: 
    Scer_pos = YGOB_pillars['Scer_' + pos]
    Scer_pos_find = Scer_pos[Scer_pos==gene]
    if len(Scer_pos_find)==1:
        if gene_pos != '':
            print('Warning: found gene in two pillars')
        gene_pos = pos
        gene_index = Scer_pos_find.index[0]
    elif len(Scer_pos_find)>1: 
        print('Warning: found gene twice in same pillar')

specs = ['Kafr','Cgla','Suva', 'Skud', 'Smik','Scer']
specs_not_scer = specs.copy()
specs_not_scer.remove('Scer')
##to include pre wgd species would need to update
specs_pillar = [spec + '_' + gene_pos for spec in specs]

orthologs = YGOB_pillars.loc[gene_index, specs_pillar]

#Extract coordinates and sequence for each ortholog
for spec_pos, gene in orthologs.iteritems():
    if gene != '---':
        spec = spec_pos.split('_')[0]
        if spec != 'Scer':
            #extract metadata for gene from spec_lookups
            gene_metadata = spec_lookups[spec].loc[gene,:]
            strand = gene_metadata['strand']
            chromosome = gene_metadata['chromosome']

            #extract sequence for chromosome: 
            fname_genome = os.path.normpath(data_processing_dir + 'promoter_phylogenies/genomes/' + spec_fullname_dict[spec] + '_sequence.fsa')
            spec_genome = SeqIO.index(fname_genome, "fasta")
            spec_chr = spec_genome[spec.lower()+'_Chr_'+str(chromosome)]

            if strand ==1: 
                fprime = gene_metadata['start']
                tprime = gene_metadata['stop']
                gene_seq = spec_chr.seq[(fprime-1):tprime]
                prom_seq = spec_chr.seq[(fprime-1-N_prom):fprime-1]
            if strand == 0: 
                fprime = gene_metadata['stop']
                tprime = gene_metadata['start']
                gene_seq = spec_chr.seq[(tprime-1):fprime].reverse_complement()
                prom_seq = spec_chr.seq[fprime:fprime+N_prom].reverse_complement()    
            prom_description = str(N_prom) + ' bases upstream of ' + gene + ' Strand: ' + strand_dict[strand] 
            prom_seq_record = SeqRecord(prom_seq,id=gene+'_prom',description = prom_description)
            prom_records.append(prom_seq_record)


output_fname = os.path.normpath(data_processing_dir+ 'promoter_phylogenies/Sensu_stricto_prom_N'+str(N_prom)+'_'+gene_name+'.fasta')
SeqIO.write(prom_records, output_fname, "fasta")
#Extract gene and promoter sequence for each ortholog. 
```
